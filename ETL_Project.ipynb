```python
url = "https://api.rawg.io/api/games"

params = {
    'key': game_API
    
    }
```

```python
platform =[]
release = []
rating = []
genres = []

# Pass the Game Name into the for loop to search for the data.... 

for row in merge_top['Game']:
    params['search'] = f"{row}"
    response = requests.get(url,params=params)
    game_info = response.json()
    
    # Fetch the Data from the site....
    
    print(f"Gather platform name for: {row}")
    pprint(f"Platform: {game_info['results'][0]['platforms'][0]['platform']['name']}")
    pprint(f"The Released Date is: {game_info['results'][0]['released']}")
    pprint(f"The Rating is: {game_info['results'][0]['rating']}")
    pprint(f"The Generes is: {game_info['results'][0]['genres'][0]['name']}")
    
    #pprint(game_info['results'][0])
    try:
        # Print the Data found from the Site using API....
        
        print("----------------------------------------------------------------------")
        print("Data Found")
        platform.append(game_info['results'][0]['platforms'][0]['platform']['name'])
        release.append(game_info['results'][0]['released'])
        rating.append(game_info['results'][0]['rating'])
        genres.append(game_info['results'][0]['genres'][0]['name'])
        print("----------------------------------------------------------------------")
    except:
        # If data not found on site append the list with NaN....
        
        print("Data Not Found")
        platform.append("NaN")
        release.append("NaN")
        rating.append("NaN")
        genres.append("NaN")
```

```python
# Create new Tables into the merge table to store the new data found....

merge_top["Platform"]  = platform
merge_top["Rating"] = rating
merge_top["Release Date"] = release
merge_top["Genres"] = genres

merge_top
```

```python
# Please do read this first...
'''Before you Run the Splinter Part please note that, I have created two different website url to scrape the and
website the obtain the data.
So, when you Run the cell Either run the Part- I or Part- II at a time...
And not the both cells simantinuosly...''' 
```

```python
#Setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# A list created to make a url to fetch data for Part- I
```

```python
table_for_url = merge_two
game_list = []
#game_list.append(m['Game'])
game_list = lambda x: x.replace(" ", "-")

game_l = list(map(game_list, table_for_url['Game']))
#game_l
[y.lower() for y in game_l]


#type(game_l)
```

```python
# Part - I
# This is Part- I, here I have used IGDB.com url to fetch the data.... 
# If you are running this part-I than please don't run Part -II....
```

```python
developer = []
image_link = []

# Pass the list created into the loop to find the game on IGDB website.... 

for b in [y.lower() for y in game_l]:    
    igdb_url = f"https://www.igdb.com/games/{b}"
    browser.visit(igdb_url)
    html = browser.html
    soup = bs(html, 'html.parser')
    time.sleep(3)
    try:
        # Fetch the Data...
        
        game_developer = soup.find('a', class_= 'block').get_text()
        game_img = soup.find('div', class_= 'media-thumbs-item media-thumbs-item-image')
        game_image = game_img.find('img')['src']
        print("--------------------------------------------------------------------")
        pprint(f" The Game Developer is: {game_developer}")
        pprint(f"The Url for Game Image is: {game_image}")
        print("--------------------------------------------------------------------")
        developer.append(game_developer)
        image_link.append(game_image)
    except:
        
        # If Data is not available append the list with NaN..
        
        print("NOT FOUND")
        developer.append("NaN")
        image_link.append("NaN")
```

```python
# Part - II 
# In this part I have used Wikipedia wwebsite to fetch the data using scrapping...
# Please Make sure you have not Run Part - I before runnig the cells below..
```

```python
developer = []
image_link = []

# Pass the Game name into the for loop to obtain the data....

for x in merge_two['Game']:
    wiki_url = f"https://en.wikipedia.org/wiki/{x}"
    browser.visit(wiki_url)
    html = browser.html
    soup = bs(html, 'html.parser')
    time.sleep(3)
    try:
        game_link = soup.find('td', class_ = 'infobox-data')
        image = soup.find('td', class_='infobox-image')
        image_url = image.find('img')['src']
        game_developer = game_link.find('a').get_text()
        print("--------------------------------------------------")
        print(f"Image url is: {image_url}")
        print(f"Game Developer is :{game_developer}")
        developer.append(game_developer)
        image_link.append(image_url)
        print("--------------------------------------------------")
    except:
        print("NOT FOUND")
        developer.append("NaN")
        image_link.append("NaN")
    
```

```python
# Quit the browser..

browser.quit()
```
